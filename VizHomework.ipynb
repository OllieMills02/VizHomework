```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

#Import the dataset into a DataFrame
df = pd.read_csv('/content/VizHomework/Top_Highest_Openings.csv')

plt.figure(figsize=(10, 6))
plt.scatter(df['Theaters'], df['Opening'], color='blue', alpha=0.5)
plt.title('Opening vs. Theaters')
plt.xlabel('Theaters')
plt.ylabel('Opening (in million)')
plt.grid(True)
plt.show()

df['% of Total'].plot(kind='hist', bins=20, title='% of Total')
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.show()

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

start_date = '2020-01-01'  # Example start date
end_date = '2022-01-01'    # Example end date
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]


plt.figure(figsize=(12, 6))
filtered_df.boxplot(column='Opening', by='Date', rot=90)
plt.title('Box Plot of Opening Gross by Date ({} to {})'.format(start_date, end_date))
plt.xlabel('Date')
plt.ylabel('Opening Gross')
plt.tight_layout()
plt.show()
```
